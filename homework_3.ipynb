```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
def vec_length(x): 
    s = 0
    for i in x:
        s += i ** 2  
    return math.sqrt(s)
 
vec_length([10,10,10])
```

```python
vec_length([0,0,-10])
```

```python
    1. Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10).Напишите код на Python, 
    реализующий их сумму.
```

```python
def vec_sum(a, b):
    if len(a) != len(b):
        return 'Error a != b'
    
    summa_vec = []
    for i in range(len(a)):
        summa_vec.append(a[i] + b[i])
    return summa_vec

vec_sum([10,10,10], [0,0,-10])
```

```python
x = []
y1 = []
y2 = []
R = 2
for i in range(1000*2*R + 1):
    x_ = -R + i/1000
    x.append(x_)
    y1.append(math.sqrt(R**2-x_**2))
    y2.append(-math.sqrt(R**2-x_**2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-3, 3)
plt.xlim(-4.3, 4.3)
plt.grid(True)
```

```python
     1. Нарисуйте график функции: для некоторых (2-3 различных) значений параметров k, a, b
        y(x) = k∙cos(x – a) + b
```

```python
def my_funct(k, a, b):
    x = np.linspace(-2*np.pi, 2*np.pi, 301)
    y = k * np.cos(x - a) + b
    plt.plot(x, np.cos(x))
    plt.plot(x, y)
    plt.grid(True)
    plt.show()
```

```python
my_funct(2, 1, 0)
my_funct(3, 0, 0)
my_funct(3, 0, -4)
```
